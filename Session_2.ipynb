CRUD
----

- C --> create

- R --> read/fetch/get

- U --> update/modify/alter

- D --> delete / drop

```python
import numpy as np
import pandas as pd
```

C --> create
---

```python
data = {
    'names' : ['Ramesh','Suresh','Satish','Alex','John'],
    'age' : [21,22,23,24,25],
    'id' : [101,102,103,104,105],
    'address' : ['aaa','bbb','ccc','ddd','eee']
    }
```

```python
df = pd.DataFrame(data)
```

```python
print(df.to_string(index=False)) # to remove index column
```
