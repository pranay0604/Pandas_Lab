```python
import pandas as pd

data = {
    'Student ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Full Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'David Lee', 'Eva Williams', 'Frank Miller', 'Grace Davis', 'Henry Wilson', 'Ivy Moore', 'Jack Taylor'],
    'Age': [18, 19, 18, 20, 19, 18, 20, 19, 18, 20],
    'Marks': [85, 92, 78, 90, 88, 75, 95, 82, 89, 79]
}

df = pd.DataFrame(data)

print(df)
```

```python
df['Age_Status'] = df['Age'].map(lambda x : "Minor" if x<=18 else "Major")
```

```python
df
```

```python
def student_age_status(age):
    if age<=18:
        return "Minor"
    else:
        return "Major"


df['Student_Age_Status'] = df['Age'].map(student_age_status)
```

```python
df
```

```python
df.map(lambda x : x*2)
```

```python
df[['Age','Marks']].map(lambda x : x*2)
```

```python
df[['Age','Marks']].apply(lambda x : x*2)
```

```python
df['Grade'] = df[['Age','Marks']].apply(lambda x : x['Age']+x['Marks'],axis=1)
```

```python
df
```

```python
def student(data):
    return pd.Series({
        'X': data['Age'] + data['Marks'],
        'Y': data['Age'] - data['Marks']
        })
```

```python
df[['X', 'Y']] = df.apply(student, axis=1)
```

```python
df
```

```python
df['Final'] = df[['Age','Marks']].apply(lambda x : "A" if (x['Age']>18 and x['Marks']>90) else "B", axis=1 )
```

```python
import pandas as pd
from datetime import datetime


data = {
    'emp id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'full name': ['Alice Smith', 'Bob Johnson', 'Charlie Williams', 'Diana Brown', 'Eve Jones',
                  'Frank Garcia', 'Grace Miller', 'Henry Davis', 'Ivy Rodriguez', 'Jack Martinez'],
    'salary': [60000, 75000, 50000, 80000, 65000, 70000, 55000, 90000, 62000, 78000],
    'department': ['HR', 'IT', 'Sales', 'IT', 'Finance', 'Sales', 'HR', 'IT', 'Finance', 'Sales'],
    'position': ['Manager', 'Engineer', 'Associate', 'Manager', 'Analyst', 'Associate', 'Analyst', 'Manager', 'Analyst', 'Engineer'],
    'hiredate': [
        datetime(2020, 1, 15), datetime(2019, 3, 10), datetime(2021, 5, 20), datetime(2018, 11, 1),
        datetime(2022, 9, 30), datetime(2020, 6, 25), datetime(2023, 2, 18), datetime(2017, 7, 7),
        datetime(2022, 4, 12), datetime(2021, 12, 1)
    ]
}

df = pd.DataFrame(data)

print(df)
```

```python
df
```

```python
df.groupby('department')['salary'].mean()
```

```python
df.groupby('position')['salary'].mean()
```

```python
df.groupby('department').count()
```

```python
df.groupby('department').size()
```

```python
df.groupby('emp id')['salary'].mean()
```

```python
df.groupby(['position','department'])['salary'].mean()
```

```python
df.groupby(['department','position'])['salary'].mean()
```

```python
df.groupby(['department','position'])['salary'].agg(['mean','max','min'])
```

```python
df.groupby(['department','position'])['salary'].agg(
    Salary_Max = 'max',
    Salary_Min = 'min',
    Salary_Mean = 'mean'
)
```
