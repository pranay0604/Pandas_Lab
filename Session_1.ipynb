Pandas
------

- The Pandas is a Python library 

- It is top on NumPy

- What are the uses of Pandas

    - data cleaning
    
    - to handle large data sets
    
    - to handle data preprocessing
    
    - to handle data analysis
    
    - to handle data manipulation
    
    - data visualisation
 
    - data transform
 
- Pandas is the process with the tabular format

- Pandas is the process with the CSV and Excel and SQL format and others

- It works well with other libraries, those are scikit-learn, OpenCV, Matplotlib and Seaborn.

- Pandas support mainly two datasets process those are :

    - Series : 1D Array
    
    - DataFrame : 2D Array or Tabular
 
- Difference between numpy and pandas :-

- NUMPY -

    - structure : It contains homogeneous with non labels
    
    - data types : homogeneous (int, float, complex, str, boolean)
    
    - data labels : it contains no labels of row and column
    
    - work : Numpy work speed and support for Statistics

- PANDAS -

    - structure : It contains tabular format data with heterogeneous 
    
    - data types : heterogeneous (mix of any object +int + float + boolean)
    
    - data labels : it contains labels of row and column
    
    - work : Pandas work speed and support for mathematics and tabular format 

Series
----

- the series contains 1D array

- by default series have labels from 0 to n.

- how to access series --> pd.Series()

```python
import pandas as pd
```

```python
pd.Series([1,2,3,4,5,6])
```

```python
pd.Series([3,'Hi',4.5,4+5j])
```

```python
data = pd.Series([3,'Hi',4.5,4+5j])
data.info()
```

```python
data = pd.Series([3,'Hi',4.5,4+5j], name='sample')
data.info()
```

```python
pd.Series({1,2,3,4,5,6})
```

```python
pd.Series((1,2,3,4,5,6))
```

```python
import numpy as np

arr = np.array([1,2,3,4,5])
pd.Series(arr)
```

```python
pd.Series(['Ram','Shyam','Laxman'])
```

```python
pd.Series(['Ram','Shyam','Laxman'],index=['A','B','C'])
```

```python
pd.Series(['Ram','Shyam','Laxman'],index=['A','B'])
```

```python
# pd.Series(data=None,
#         index=None,
#         dtype: 'Dtype | None' = None,
#         name=None)
```

```python
pd.Series([1,2,3,4,5],dtype='str')
```

```python
pd.Series([1,2,3,4,5],dtype='float')
```

```python
pd.Series([1,2,3,4,5],dtype='bool')
```

DataFrame
----

- It is 2D array / tabular format

- It contains rows --> records or tuples or values and columns --> fields or attributes or labels

- it contains index and column

```python
pd.DataFrame([1,2,3,4])
```

```python
pd.DataFrame([1,2,3,4],columns=['A'])
```

```python
pd.DataFrame([1,2,3,4],index=['a','b','c','d'] ,columns=['A'])
```

```python
pd.DataFrame([[1,2,3,4],[5,6,7,8]],columns=['A','B','C','D'],index=[101,102])
```

```python
pd.DataFrame({'Name':['Ram','Shyam','Laxman'],
              'Age':[22,24,26]})
```

```python
# dictionary --> key value pair
# keys : columns / labels
# values : rows / records
```

```python
pd.DataFrame({'Name':{'Ram','Shyam','Laxman'},
              'Age':[22,24,26]})
```

```python
pd.DataFrame({ 'Name':np.array(['Ram','Shyam','Laxman']),
              'Age':np.array([22,24,26])
             })
```

```python
pd.DataFrame({ 'Name':np.array(['Ram','Shyam','Laxman']),
              'Age':range(3)
             })
```

```python
pd.DataFrame({ 'Name':np.array(['Ram','Shyam','Laxman']),
              'Age':np.arange(22,27,2)
             })
```

```python
data = pd.DataFrame({ 'Name':np.array(['Ram','Shyam','Laxman']),
              'Age':np.arange(22,27,2)
             })

data.head(2)
```
